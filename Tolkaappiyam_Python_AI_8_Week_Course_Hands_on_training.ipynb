```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
தொல்காப்பியம் - எழுத்துப் பிறப்பியல் விளையாட்டு
Tolkappiyam - Tamil Phonetics Learning Game

தொல்காப்பியத்தின் நூற்பாக்கள் 83-103 அடிப்படையில்,
தமிழ் உயிர், மெய், ஆய்த எழுத்துகளின் பிறப்பிடங்களை
ஊடாடும் விளையாட்டாக வடிவமைத்தல்.
"""

import random
import sys
from typing import Dict, List, Tuple, Optional

# ============================================================
# தொல்காப்பிய எழுத்துப் பிறப்பியல் தரவுகள்
# ============================================================

# உயிரெழுத்துகள் - பிறப்பிட விளக்கம் (நூற்பா 83-87)
UYIR_PIRAPPIDAM = {
    'அ': 'உந்தி முதலா முந்துவளி தோன்றி - தொண்டைக் குழியிலிருந்து',
    'ஆ': 'உந்தி முதலா முந்துவளி தோன்றி - தொண்டைக் குழியிலிருந்து (நீண்ட)',
    'இ': 'நுனிநா அண்ணம் - நாக்கு மேலண்ணத்தைத் தொட',
    'ஈ': 'நுனிநா அண்ணம் - நாக்கு மேலண்ணத்தைத் தொட (நீண்ட)',
    'எ': 'இடைநா அண்ணம் - நாக்கு மேலண்ணத்தை அணுக',
    'ஏ': 'இடைநா அண்ணம் - நாக்கு மேலண்ணத்தை அணுக (நீண்ட)',
    'ஐ': 'இஈ எஏ ஐ என இசைக்கும் - கூடிய உயிர்',
    'உ': 'இதழ் குவிந்து - இதழ்களைக் குவித்து',
    'ஊ': 'இதழ் குவிந்து - இதழ்களைக் குவித்து (நீண்ட)',
    'ஒ': 'இதழ் குவிந்து - நாக்கு பின் செல்ல',
    'ஓ': 'இதழ் குவிந்து - நாக்கு பின் செல்ல (நீண்ட)',
    'ஔ': 'உஊ ஒஓ ஔ என இசைக்கும் - கூடிய உயிர்'
}

# மெய்யெழுத்துகள் - பிறப்பிட விளக்கம் (நூற்பா 89-100)
MEY_PIRAPPIDAM = {
    'க்': 'ககாரம் - முதல் நா அண்ணம்',
    'ங்': 'ஙகாரம் - முதல் நா அண்ணம் (மெல்லிய)',
    'ச்': 'சகாரம் - இடை நா அண்ணம்',
    'ஞ்': 'ஞகாரம் - இடை நா அண்ணம் (மெல்லிய)',
    'ட்': 'டகாரம் - நுனி நா அண்ணம்',
    'ண்': 'ணகாரம் - நுனி நா அண்ணம் (மெல்லிய)',
    'த்': 'தகாரம் - பல் + நுனிநா அண்ணம்',
    'ந்': 'நகாரம் - பல் + நுனிநா அண்ணம் (மெல்லிய)',
    'ப்': 'பகாரம் - இதழியைந்து',
    'ம்': 'மகாரம் - இதழியைந்து (மெல்லிய)',
    'ய்': 'யகாரம் - அண்ணஞ் சேர்ந்த மிடற்றெழு வளி',
    'ர்': 'ரகாரம் - நுனிநா அணரியண்ணம்',
    'ல்': 'லகாரம் - நாவிளிம்பு வீங்கி',
    'வ்': 'வகாரம் - பல்லிதழியைய',
    'ழ்': 'ழகாரம் - நுனிநா வணரியண்ணம்',
    'ள்': 'ளகாரம் - நாவிளிம்பு வீங்கி',
    'ற்': 'றகாரம் - அணரிநுனி அண்ணம்',
    'ன்': 'னகாரம் - அணரிநுனி அண்ணம் (மெல்லிய)'
}

# ஆய்த எழுத்து (நூற்பா 94)
AAYTHAM = {
    'ஃ': 'ஃகான் - அணரிநுனி வளி முனைப்பு'
}

# முழு எழுத்துப் பட்டியல்
ALL_LETTERS = {
    'உயிர்': ['அ', 'ஆ', 'இ', 'ஈ', 'உ', 'ஊ', 'எ', 'ஏ', 'ஐ', 'ஒ', 'ஓ', 'ஔ'],
    'மெய்': ['க்', 'ங்', 'ச்', 'ஞ்', 'ட்', 'ண்', 'த்', 'ந்', 'ப்', 'ம்', 'ய்', 'ர்', 'ல்', 'வ்', 'ழ்', 'ள்', 'ற்', 'ன்'],
    'ஆய்தம்': ['ஃ']
}

# எழுத்துக்களை வகைப்படுத்தல்
LETTER_TYPE = {}
for letter in ALL_LETTERS['உயிர்']:
    LETTER_TYPE[letter] = 'உயிர்'
for letter in ALL_LETTERS['மெய்']:
    LETTER_TYPE[letter] = 'மெய்'
for letter in ALL_LETTERS['ஆய்தம்']:
    LETTER_TYPE[letter] = 'ஆய்தம்'

# ============================================================
# விளையாட்டு வகுப்பு
# ============================================================

class TolkappiyamGame:
    """தொல்காப்பிய எழுத்துப்பிறப்பியல் விளையாட்டு"""

    def __init__(self):
        self.score = 0
        self.total_questions = 0
        self.difficulty = 'normal'  # easy, normal, hard
        self.hints_used = 0

    def display_header(self):
        """விளையாட்டுத் தலைப்புக் காட்டி"""
        print("\n" + "="*60)
        print("    🕉️  தொல்காப்பியம் - எழுத்துப் பிறப்பியல் விளையாட்டு  🕉️")
        print("="*60)
        print("    நூற்பாக்கள் 83-103 | எழுத்துகளின் பிறப்பிடம் அறிதல்")
        print("="*60)
        print("\n📜 \"எல்லா வெழுத்தும் வெளிப்படக்கிளந்து...\"")
        print("   - தொல்காப்பியம், எழுத்ததிகாரம்\n")

    def display_nurpa(self, nurpa_num: int):
        """தொல்காப்பிய நூற்பாவைக் காட்டு"""
        nurpas = {
            83: """நூற்பா 83: "உந்தி முதலா முந்துவளி தோன்றித்
            தலையினு மிடற்றினு நெஞ்சினு நிலைஇப்
            பல்லு மிதழு நாவு மூக்கு மண்ணமு முளப்பட
            வெண்முறை நிலையானுறுப்புற் றமைய..." """,
            89: """நூற்பா 89: "ககார ஙகார முதனா வண்ணம்" """,
            90: """நூற்பா 90: "சகார ஞகார மிடைநா வண்ணம்" """,
            91: """நூற்பா 91: "டகார ணகார நுனிநா வண்ணம்" """,
            97: """நூற்பா 97: "இதழியைந்து பிறக்கும் பகார மகாரம்" """,
            98: """நூற்பா 98: "பல்லிதழியைய வகாரம் பிறக்கும்" """
        }
        if nurpa_num in nurpas:
            print(f"\n📖 {nurpas[nurpa_num]}\n")

    def get_question(self) -> Tuple[str, str, str]:
        """சீரற்ற எழுத்து மற்றும் கேள்வி வகையைத் தேர்வு செய்"""
        if self.difficulty == 'easy':
            letters = ALL_LETTERS['உயிர்'][:6] + ALL_LETTERS['மெய்'][:8]
        elif self.difficulty == 'hard':
            letters = ALL_LETTERS['உயிர்'] + ALL_LETTERS['மெய்'] + ALL_LETTERS['ஆய்தம்']
        else:
            letters = ALL_LETTERS['உயிர்'] + ALL_LETTERS['மெய்']

        letter = random.choice(letters)
        letter_type = LETTER_TYPE[letter]

        # கேள்வி வகை: 0=பிறப்பிடம் கேட்க, 1=எழுத்து கேட்க
        q_type = random.choice([0, 1])

        if q_type == 0:
            question = f"🔊 '{letter}' என்ற எழுத்து எங்கிருந்து பிறக்கும்?"
            if letter in UYIR_PIRAPPIDAM:
                answer = UYIR_PIRAPPIDAM[letter]
            elif letter in MEY_PIRAPPIDAM:
                answer = MEY_PIRAPPIDAM[letter]
            else:
                answer = AAYTHAM[letter]
            hint = self._get_hint(letter)
        else:
            # பிறப்பிடம் தந்து எழுத்து கேட்க
            all_answers = list(UYIR_PIRAPPIDAM.keys()) + list(MEY_PIRAPPIDAM.keys())
            if self.difficulty == 'hard':
                all_answers += list(AAYTHAM.keys())
            correct = random.choice(all_answers)
            desc = UYIR_PIRAPPIDAM.get(correct) or MEY_PIRAPPIDAM.get(correct) or AAYTHAM.get(correct, "")
            question = f"📍 '{desc[:40]}...' என்ற பிறப்பிடத்தை உடைய எழுத்து எது?"
            answer = correct
            hint = f"💡 இந்த எழுத்து {LETTER_TYPE[correct]} வகையைச் சேர்ந்தது."

        return question, answer, hint

    def _get_hint(self, letter: str) -> str:
        """எழுத்துக்கான குறிப்பு"""
        if letter in UYIR_PIRAPPIDAM:
            return f"💡 இது ஓர் உயிரெழுத்து. தொல்காப்பியம் 12 உயிர்களை {len(ALL_LETTERS['உயிர்'])} வகையாகப் பிரிக்கும்."
        elif letter in MEY_PIRAPPIDAM:
            # மெய்யின் வகை காணல்
            if letter in ['க்', 'ங்']:
                return "💡 இது முதல் நா அண்ண மெய் (நூற்பா 89)"
            elif letter in ['ச்', 'ஞ்']:
                return "💡 இது இடை நா அண்ண மெய் (நூற்பா 90)"
            elif letter in ['ட்', 'ண்']:
                return "💡 இது நுனி நா அண்ண மெய் (நூற்பா 91)"
            elif letter in ['ப்', 'ம்']:
                return "💡 இது இதழியைந்த மெய் (நூற்பா 97)"
            elif letter == 'வ்':
                return "💡 இது பல்லிதழியைய மெய் (நூற்பா 98)"
            else:
                return "💡 இது வல்லினம்/மெல்லினம்/இடையின மெய்யாக இருக்கலாம்."
        else:
            return "💡 இது ஆய்த எழுத்து. 'அணரிநுனி வளி முனைப்பு' (நூற்பா 94)"

    def check_answer(self, user_input: str, correct: str) -> bool:
        """விடையைச் சரிபார்"""
        return user_input.strip() == correct.strip()

    def play_round(self):
        """ஒரு சுற்று விளையாடு"""
        question, answer, hint = self.get_question()

        print(f"\n{'─'*50}")
        print(f"📝 கேள்வி {self.total_questions + 1}:")
        print(f"{question}")
        print(f"{'─'*50}")

        # உதவி விருப்பம்
        print("\n[1] விடை தருக  [2] உதவி/குறிப்பு  [3] விடை தெரியவில்லை")
        choice = input("\nதேர்வு செய்க (1/2/3): ").strip()

        if choice == '2':
            print(f"\n{hint}")
            self.hints_used += 1
            user_answer = input("\nஇப்போது உங்கள் விடை: ").strip()
        elif choice == '3':
            print(f"\n📖 சரியான விடை: {answer}")
            print("⭐ தொல்காப்பியம் கூறும் பிறப்பியல் வழியே தொடர்ந்து கற்போம்!")
            return False
        else:
            user_answer = input("\nஉங்கள் விடை: ").strip()

        if self.check_answer(user_answer, answer):
            points = 10
            print(f"\n✅ சரியான விடை! (+{points} புள்ளிகள்)")
            print(f"📚 {answer}")
            self.score += points
            return True
        else:
            print(f"\n❌ முறையான விடை அல்ல.")
            print(f"📖 சரியான விடை: {answer}")
            print(f"📚 விளக்கம்: {answer}")
            return False

    def display_progress(self):
        """முன்னேற்றத்தைக் காட்டு"""
        print(f"\n{'─'*40}")
        print(f"🏆 மதிப்பெண்: {self.score}")
        print(f"📊 கேள்விகள்: {self.total_questions}")
        if self.total_questions > 0:
            accuracy = (self.score / (self.total_questions * 10)) * 100
            print(f"📈 துல்லியம்: {accuracy:.1f}%")
        print(f"💡 உதவிகள் பயன்படுத்தியவை: {self.hints_used}")
        print(f"{'─'*40}")

    def show_learning_table(self):
        """கற்றல் அட்டவணை காட்டு"""
        print("\n" + "="*60)
        print("📖 தொல்காப்பிய எழுத்துப் பிறப்பியல் சுருக்கம்")
        print("="*60)

        print("\n🔴 உயிரெழுத்துகள் (12):")
        print("   அ-ஆ : தொண்டைக் குழி | இ-ஈ : நுனிநா அண்ணம்")
        print("   உ-ஊ : இதழ் குவிந்து | எ-ஏ : இடைநா அண்ணம்")
        print("   ஐ, ஔ : கூடிய உயிர்கள்")

        print("\n🟢 மெய்யெழுத்துகள் (18):")
        print("   க், ங் : முதல் நா அண்ணம் (நூற்பா 89)")
        print("   ச், ஞ் : இடை நா அண்ணம் (நூற்பா 90)")
        print("   ட், ண் : நுனி நா அண்ணம் (நூற்பா 91)")
        print("   த், ந் : பல் + நுனிநா (நூற்பா 93)")
        print("   ப், ம் : இதழியைந்து (நூற்பா 97)")
        print("   ர், ழ், ல், ள், ற், ன், ய், வ் : இடையினம்")

        print("\n🟡 ஆய்தம்:")
        print("   ஃ : அணரிநுனி வளி முனைப்பு (நூற்பா 94)")

        print("\n" + "="*60)

    def run(self):
        """முதன்மை விளையாட்டு இயக்கி"""
        self.display_header()

        # கடினத்தேர்வு
        print("\n🎮 கடினத்தேர்வு:")
        print("   [1] இளையர் நிலை (எளிது)")
        print("   [2] பொது நிலை (இயல்பு)")
        print("   [3] புலவர் நிலை (கடினம்)")
        diff_choice = input("\nதேர்வு: ").strip()

        if diff_choice == '1':
            self.difficulty = 'easy'
            rounds = 10
        elif diff_choice == '3':
            self.difficulty = 'hard'
            rounds = 20
        else:
            self.difficulty = 'normal'
            rounds = 15

        print(f"\n✨ {rounds} கேள்விகள் விளையாடப் போகிறோம்...")
        input("\nதொடங்க Enter அழுத்தவும் ▶️")

        # கற்றல் அட்டவணை
        print("\n📚 முதலில் அடிப்படைப் பாடத்தைக் காணலாமா? (y/n)")
        if input().lower() == 'y':
            self.show_learning_table()
            input("\nதயாரானதும் Enter அழுத்தவும்...")

        # விளையாட்டுச் சுழற்சி
        for round_num in range(rounds):
            self.total_questions += 1

            # ஒவ்வொரு 5 கேள்விகளுக்கும் நூற்பா காட்டு
            if self.total_questions % 5 == 0:
                nurpa_num = random.choice([83, 89, 90, 91, 97, 98])
                self.display_nurpa(nurpa_num)

            self.play_round()
            self.display_progress()

            if self.total_questions < rounds:
                cont = input("\nதொடரவா? (Enter - தொடர, q - வெளியேறு): ").strip()
                if cont.lower() == 'q':
                    break

        # இறுதி முடிவு
        print("\n" + "="*60)
        print("🏁 விளையாட்டு முடிந்தது!")
        print("="*60)
        self.display_progress()

        # பாராட்டு
        if self.score >= rounds * 8:
            print("\n🌟 அருமை! நீங்கள் ஒரு தொல்காப்பிய மாணவர்!")
        elif self.score >= rounds * 5:
            print("\n📖 நல்ல முயற்சி! தொடர்ந்து பயிலுங்கள்!")
        else:
            print("\n🌱 தொடர்ந்து கற்றல் மூலம் மேலும் தேர்ச்சி பெறலாம்!")

        print("\n🙏 தொல்காப்பியர் அருளிய எழுத்திலக்கணம் போற்றிப் புகழ்வோம்!")
        print("="*60)


# ============================================================
# நிரலை இயக்கும் இடம்
# ============================================================

def main():
    game = TolkappiyamGame()
    try:
        game.run()
    except KeyboardInterrupt:
        print("\n\n👋 விடைபெறுகிறேன். நன்றி!")
        sys.exit(0)

if __name__ == "__main__":
    main()
```

```python
# நூற்பா:01 (எழுத்தெனப்படுப) - Tolkappiyam Sutra: 01 (Regarding Letters)

# இளம்பூரணர் உரை விளக்கத்தின்படி தமிழ் எழுத்துக்களின் வகைப்பாடுகள், தொகைகள்:

# முதல் எழுத்துக்கள் (Primary Letters)
# உயிர் எழுத்துக்கள் (Vowels)
uyir_eluthukkal = ['அ', 'ஆ', 'இ', 'ஈ', 'உ', 'ஊ', 'எ', 'ஏ', 'ஐ', 'ஒ', 'ஓ', 'ஔ']
print(f"உயிர் எழுத்துக்கள் (Vowels): {', '.join(uyir_eluthukkal)} (மொத்தம்: {len(uyir_eluthukkal)})")

# மெய் எழுத்துக்கள் (Consonants)
mei_eluthukkal = ['க்', 'ங்', 'ச்', 'ஞ்', 'ட்', 'ண்', 'த்', 'ந்', 'ப்', 'ம்', 'ய்', 'ர்', 'ல்', 'வ்', 'ழ்', 'ள்', 'ற்', 'ன்']
print(f"மெய் எழுத்துக்கள் (Consonants): {', '.join(mei_eluthukkal)} (மொத்தம்: {len(mei_eluthukkal)})")

# முப்பஃது என்ப - முதல் எழுத்துக்களின் மொத்த எண்ணிக்கை
muthal_eluthukkal_count = len(uyir_eluthukkal) + len(mei_eluthukkal)
print(f"\nமுதல் எழுத்துக்கள் (Primary Letters): {muthal_eluthukkal_count} ('முப்பஃது என்ப')")

# சார்பெழுத்துக்கள் (Dependent Letters) - சார்ந்துவரல் மரபின் மூன்று
sarbu_eluthukkal = ['ஆய்தம்', 'குற்றியலிகரம்', 'குற்றியலுகரம்'] # இளம்பூரணர் உரையில் வெளிப்படையாகக் குறிப்பிடப்படவில்லை, ஆனால் தொல்காப்பிய மரபுப்படி இவை சார்பெழுத்துக்கள்.
print(f"சார்பெழுத்துக்கள் (Dependent Letters): {', '.join(sarbu_eluthukkal)} (மொத்தம்: {len(sarbu_eluthukkal)})")

# எழுத்துக்களின் மொத்த எண்ணிக்கை - வெவ்வேறு நிலைகள்
print("\nஎழுத்துக்களின் வகைப்பாடுகள் மற்றும் தொகைகள்:")

# 1. முதல் எழுத்துக்கள் மட்டும்
count_stage_1 = muthal_eluthukkal_count
print(f"1. முதல் எழுத்துக்கள் மட்டும்: {count_stage_1}")

# 2. முதல் எழுத்துக்கள் + சார்பெழுத்துக்கள் (மூன்றும் ஆனவிடத்து முப்பத்து மூன்று)
count_stage_2 = muthal_eluthukkal_count + len(sarbu_eluthukkal)
print(f"2. முதல் எழுத்துக்கள் + சார்பெழுத்துக்கள்: {count_stage_2}")

# 3. அளபெடை உள்ளிட்டவை (நாற்பது)
# 'அளபெடை' என்பது ஒரு குறிப்பிட்ட எண்ணிக்கையிலான எழுத்துக்கள் அல்லாமல், உயிரெழுத்துக்கள் நீண்டு ஒலிக்கும் ஒரு இலக்கணக் கூற்று.
# உரையின்படி இது கணக்கில் கொள்ளப்படும்போது மொத்த எண்ணிக்கை 40 ஆகிறது.
count_stage_3 = 40 # 'அளபெடை தலைப்பெய்து நாற்பதென்பது அதன் விரி.'
print(f"3. அளபெடை உள்ளிட்டவை (விரி): {count_stage_3}")

# 4. உயிர்மெய் உள்ளிட்டவை (இருநூற்றைம்பத்தாறு)
# 'உயிர்மெய்' என்பது உயிர் எழுத்துக்களும் மெய் எழுத்துக்களும் சேர்ந்து உருவாகும் எழுத்துக்கள் (12 * 18 = 216).
# உரையின்படி இவை கணக்கில் கொள்ளப்படும்போது மொத்த எண்ணிக்கை 256 ஆகிறது.
count_stage_4 = 256 # 'உயிர்மெய் தலைப்பெய்து இருநூற்றைம்பத்தாறென்பது அதன் விரி.'
print(f"4. உயிர்மெய் உள்ளிட்டவை (விரி): {count_stage_4}")

# 'அகரம் முதல் னகர இறுவாய்' விளக்கம்
print("\n'அகரம் முதல் னகர இறுவாய்' விளக்கம் (இளம்பூரணர் உரையிலிருந்து):")
print("- அகரம் (அ): தானும் இயங்கித் தனிமெய்களை இயக்குதற் சிறப்பான் முன்வைக்கப்பட்டது.")
print("- னகரம் (ன்): வீடு பேற்றிற்குரிய ஆண்பாலை உணர்த்துதற் சிறப்பான் பின்வைக்கப்பட்டது.")
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
தொல்காப்பிய மொழியிறுதி விதிகள் - கற்றல் விளையாட்டு
Tholkaappiar Word Ending Rules - Learning Game for Tamil Literature Students

இந்நிரல் தொல்காப்பிய மொழியிறுதி விதிகளை விளையாட்டு வடிவில் கற்றுத் தருகிறது.
இளங்கலைத் தமிழ் இலக்கிய மாணவர்களை மையமாகக் கொண்டு வடிவமைக்கப்பட்டுள்ளது.

ஆசிரியர்கள்: அ.பரமேசுவரன், முனைவர் த. சத்தியராஜ் (நேயக்கோ)
"""

import tamil
import random
import sys
from typing import Dict, List, Tuple, Optional

# ============================================================
# தொல்காப்பிய மொழியிறுதி விதிகள் - தரவுகள்
# ============================================================

# விதிகளின் விளக்கம் (மாணவர்களுக்கான தெளிவான விளக்கம்)
RULE_EXPLANATIONS = {
    'uyir': {
        'name': 'உயிரெழுத்தில் முடிதல் (Uyir Ending)',
        'rule': 'சொல்லின் இறுதியில் உயிரெழுத்து (அ, ஆ, இ, ஈ, உ, ஊ, எ, ஏ, ஐ, ஒ, ஓ, ஔ) வரும்போது',
        'example': 'மரம் + அ = மரமா (மரம் + ஆ) - இங்கு "ஆ" உயிரெழுத்து',
        'tholkaappiyam': 'தொல்காப்பியம் - எழுத்ததிகாரம்: "உயிர் ஈறு ஆகிய எழுத்து"'
    },
    'mellinam': {
        'name': 'மெல்லினத்தில் முடிதல் (Mellinam Ending)',
        'rule': 'சொல்லின் இறுதியில் மெல்லின மெய் (ஞ், ண், ந், ம், ன்) வரும்போது',
        'example': '"மரம்" - "ம்" மெல்லினம்',
        'tholkaappiyam': 'தொல்காப்பியம் - மெய்யீற்று விதிகள்'
    },
    'idaiyinam': {
        'name': 'இடையினத்தில் முடிதல் (Idaiyinam Ending)',
        'rule': 'சொல்லின் இறுதியில் இடையின மெய் (ய், ர், ல், வ், ழ், ள்) வரும்போது',
        'example': 'செய், கால், தேள் - இவற்றின் இறுதி இடையினம்',
        'tholkaappiyam': 'தொல்காப்பியம் - இடையின விதிகள்'
    },
    'oorezhuthu': {
        'name': 'ஓரெழுத்தொருமொழி (Single Letter Word)',
        'rule': 'நெடில் உயிரெழுத்து (ஆ, ஈ, ஊ, ஏ, ஐ, ஓ) ஒன்றே சொல்லாக வரும்போது',
        'example': 'ஆ! (ஆமாம்), ஏ! (ஏன்), ஓ! (ஓடு)',
        'tholkaappiyam': 'தொல்காப்பியம் - "ஓரெழுத்தொருமொழி நெடில் வருமே"'
    },
    'alapedai': {
        'name': 'அளபெடை (Alapedai)',
        'rule': 'குறில் உயிரெழுத்து (அ, இ, உ, எ, ஒ) நீண்டு ஒலித்தல்',
        'example': 'அம்மா (ம்மா - மீது அளபெடை)',
        'tholkaappiyam': 'தொல்காப்பியம் - அளபெடையியல்'
    },
    'suttu': {
        'name': 'சுட்டெழுத்து (Suttu - Demonstrative)',
        'rule': 'சுட்டு வினைமுற்றில் குறில் (அ,இ,உ) + மெய் (க்,ச்,த்,ப்,வ்)',
        'example': 'அக்கொலை, இச்செயல், உவ்வழி',
        'tholkaappiyam': 'தொல்காப்பியம் - சுட்டியல்'
    },
    'vinaa': {
        'name': 'வினாச்சொல் (Interrogative)',
        'rule': 'வினாக்குறியில் "எ" + மெய் (க்,ச்,த்,ப்,வ்,ங்,ந்)',
        'example': 'எக்காலம்? எச்செய்தி? எத்தனை?',
        'tholkaappiyam': 'தொல்காப்பியம் - வினாயியல்'
    }
}

# எடுத்துக்காட்டுச் சொற்கள்
EXAMPLE_WORDS = {
    'uyir': ['மரமா', 'படி', 'கடை', 'தேரே', 'வந்தான்', 'சென்றாள்', 'பேசு', 'தேடி', 'கொண்டு', 'ஓடு'],
    'mellinam': ['வேங்கை', 'அரண்', 'கடிதம்', 'மரம்', 'தேன்', 'காண்'],
    'idaiyinam': ['செய்', 'கால்', 'வேல்', 'பால்', 'தேள்', 'மேல்', 'தீர்', 'மெய்'],
    'oorezhuthu': ['ஆ', 'ஈ', 'ஊ', 'ஏ', 'ஐ', 'ஓ'],
    'alapedai': ['அம்மா', 'இல்லை', 'உத்தரவு', 'எழில்', 'ஒப்பு'],
    'suttu': ['அக்கொலை', 'இச்செயல்', 'உவ்வழி', 'அப்பொருள்', 'இவ்வுலகு'],
    'vinaa': ['எக்காலம்', 'எச்செய்தி', 'எத்தனை', 'எப்பொருள்', 'எவ்வழி']
}

# ============================================================
# தொல்காப்பிய விதி சோதனைச் செயல்பாடுகள்
# ============================================================

def uyir_check(word):
    """சொல்லின் இறுதி உயிரெழுத்தில் முடிகிறதா?"""
    uyir_list = ["அ", "ஆ", "இ", "ஈ", "உ", "ஊ", "எ", "ஏ", "ஐ", "ஒ", "ஓ", "ஔ"]
    try:
        letters = tamil.utf8.get_letters(word)
        if not letters:
            return False
        last_letter = letters[-1]
        root_words = tamil.utf8.splitMeiUyir(last_letter)
        if isinstance(root_words, tuple):
            root_last = root_words[-1]
        else:
            root_last = root_words
        return root_last in uyir_list
    except:
        return False

def mellinam_check(word):
    """சொல்லின் இறுதி மெல்லினமாக உள்ளதா?"""
    mellinam_list = ["ஞ்", "ண்", "ந்", "ம்", "ன்"]
    try:
        letters = tamil.utf8.get_letters(word)
        if not letters:
            return False
        last_letter = letters[-1]
        root_words = tamil.utf8.splitMeiUyir(last_letter)
        if isinstance(root_words, tuple):
            root_last = root_words[-1]
        else:
            root_last = root_words
        return root_last in mellinam_list
    except:
        return False

def idaiyinam_check(word):
    """சொல்லின் இறுதி இடையினமாக உள்ளதா?"""
    idaiyinam_list = ["ய்", "ர்", "ல்", "வ்", "ழ்", "ள்"]
    try:
        letters = tamil.utf8.get_letters(word)
        if not letters:
            return False
        last_letter = letters[-1]
        root_words = tamil.utf8.splitMeiUyir(last_letter)
        if isinstance(root_words, tuple):
            root_last = root_words[-1]
        else:
            root_last = root_words
        return root_last in idaiyinam_list
    except:
        return False

def oorezhuthoorumozhi_check(word):
    """ஓரெழுத்தொருமொழியா?"""
    oorezhuthoorumozhi_list = ["ஆ", "ஈ", "ஊ", "ஏ", "ஐ", "ஓ"]
    try:
        letters = tamil.utf8.get_letters(word)
        if len(letters) == 1 and letters[0] in oorezhuthoorumozhi_list:
            return True
        return False
    except:
        return False

def alapedai_check(word):
    """அளபெடை உள்ளதா?"""
    alapedai_list = ["அ", "இ", "உ", "எ", "ஒ"]
    try:
        letters = tamil.utf8.get_letters(word)
        for i in range(len(letters)-1):
            # மெய் எழுத்து வகைகள்
            mei_list = ['க்', 'ங்', 'ச்', 'ஞ்', 'ட்', 'ண்', 'த்', 'ந்', 'ப்', 'ம்', 'ய்', 'ர்', 'ல்', 'வ்', 'ழ்', 'ள்', 'ற்', 'ன்']
            if letters[i] in mei_list:
                if i+1 < len(letters):
                    next_letter = letters[i+1]
                    if next_letter in alapedai_list:
                        return True
        return False
    except:
        return False

def suttu_check(word):
    """சுட்டெழுத்து அமைப்பா?"""
    suttu_list = ["அ", "இ", "உ"]
    suttu_mey_list = ["க்", "ச்", "த்", "ப்", "வ்"]
    try:
        letters = tamil.utf8.get_letters(word)
        if len(letters) >= 2:
            first_letter = letters[0]
            second_letter = letters[1]
            return first_letter in suttu_list and second_letter in suttu_mey_list
        return False
    except:
        return False

def vinaa_check(word):
    """வினாச்சொல்லா?"""
    vinaa_mey_list = ["க்", "ச்", "த்", "ப்", "வ்", "ங்", "ந்"]
    try:
        letters = tamil.utf8.get_letters(word)
        if len(letters) >= 2:
            first_letter = letters[0]
            second_letter = letters[1]
            return first_letter == "எ" and second_letter in vinaa_mey_list
        return False
    except:
        return False

def get_all_rules_for_word(word):
    """ஒரு சொல்லுக்கு பொருந்தும் அனைத்து விதிகளையும் கண்டறிதல்"""
    rules_matched = []
    if uyir_check(word):
        rules_matched.append('uyir')
    if mellinam_check(word):
        rules_matched.append('mellinam')
    if idaiyinam_check(word):
        rules_matched.append('idaiyinam')
    if oorezhuthoorumozhi_check(word):
        rules_matched.append('oorezhuthu')
    if alapedai_check(word):
        rules_matched.append('alapedai')
    if suttu_check(word):
        rules_matched.append('suttu')
    if vinaa_check(word):
        rules_matched.append('vinaa')
    return rules_matched

# ============================================================
# விளையாட்டு வகுப்பு
# ============================================================

class TholkaappiyamGame:
    """தொல்காப்பிய மொழியிறுதி விதிகள் கற்றல் விளையாட்டு"""

    def __init__(self):
        self.score = 0
        self.total_questions = 0
        self.streak = 0
        self.max_streak = 0
        self.correct_answers = 0

    def display_header(self):
        """விளையாட்டுத் தலைப்பு மற்றும் அறிமுகம்"""
        print("\n" + "=" * 70)
        print("🕉️  தொல்காப்பிய மொழியிறுதி விதிகள் - கற்றல் விளையாட்டு  🕉️")
        print("=" * 70)
        print("\n📜 \"எழுத்து இறுதி விதிகள் அறிதல் தொல்காப்பிய நெறி\"")
        print("=" * 70)
        print("\n🎓 இளங்கலைத் தமிழிலக்கிய மாணவர்களுக்கு")
        print("📚 தொல்காப்பியரின் மொழியிறுதி விதிகளை விளையாட்டாகக் கற்போம்!\n")

    def display_rules_summary(self):
        """விதிகளின் சுருக்கத்தைக் காட்டு"""
        print("\n" + "─" * 70)
        print("📖 தொல்காப்பிய மொழியிறுதி விதிகள் - விளக்கம்")
        print("─" * 70)

        for key, rule in RULE_EXPLANATIONS.items():
            print(f"\n🔸 {rule['name']}")
            print(f"   📌 {rule['rule']}")
            print(f"   💡 எடுத்துக்காட்டு: {rule['example']}")
            print(f"   🕉️ {rule['tholkaappiyam']}")
        print("\n" + "─" * 70)

    def get_question_type(self):
        """கேள்வி வகையைத் தேர்வு செய்"""
        q_types = ['identify_rule', 'find_example', 'true_false', 'fill_blank']
        return random.choice(q_types)

    def generate_question_identify_rule(self):
        """சொல்லைக் கொடுத்து விதியை அடையாளம் காட்டும் கேள்வி"""
        # சோதனைக்கான சொற்கள்
        test_words = [
            ('மரமா', 'uyir'), ('படி', 'uyir'), ('கடை', 'uyir'),
            ('மரம்', 'mellinam'), ('தேன்', 'mellinam'), ('அரண்', 'mellinam'),
            ('செய்', 'idaiyinam'), ('கால்', 'idaiyinam'), ('வேல்', 'idaiyinam'),
            ('ஆ', 'oorezhuthu'), ('ஏ', 'oorezhuthu'), ('ஓ', 'oorezhuthu'),
            ('அம்மா', 'alapedai'), ('இல்லை', 'alapedai'),
            ('அக்கொலை', 'suttu'), ('இச்செயல்', 'suttu'),
            ('எக்காலம்', 'vinaa'), ('எச்செய்தி', 'vinaa')
        ]

        word, correct_rule = random.choice(test_words)
        return {
            'question': f"'{word}' என்ற சொல் எந்த விதியின் கீழ் வரும்?",
            'answer': correct_rule,
            'word': word,
            'type': 'identify_rule'
        }

    def generate_question_find_example(self):
        """விதியைக் கொடுத்து எடுத்துக்காட்டுக் கேட்கும் கேள்வி"""
        rules = list(RULE_EXPLANATIONS.keys())
        selected_rule = random.choice(rules)
        return {
            'question': f"'{RULE_EXPLANATIONS[selected_rule]['name']}' விதிக்கு ஏற்ற ஒரு சொல்லை எடுத்துக்காட்டுக:",
            'answer': selected_rule,
            'type': 'find_example',
            'rule_key': selected_rule
        }

    def generate_question_true_false(self):
        """சரி/தவறு கேள்விகள்"""
        statements = [
            ("'மரம்' என்ற சொல் மெல்லினத்தில் முடிகிறது", True, 'mellinam'),
            ("'செய்' என்ற சொல் உயிரெழுத்தில் முடிகிறது", False, 'idaiyinam'),
            ("'ஆ' என்பது ஓரெழுத்தொருமொழி", True, 'oorezhuthu'),
            ("'அக்கொலை' சுட்டெழுத்து விதியில் வரும்", True, 'suttu'),
            ("'எக்காலம்' வினாச்சொல் விதியில் வரும்", True, 'vinaa'),
            ("'படி' என்ற சொல் இடையினத்தில் முடிகிறது", False, 'uyir'),
            ("'தேன்' உயிரெழுத்தில் முடிகிறது", False, 'mellinam'),
            ("'ஏ' என்பது அளபெடைக்கு எடுத்துக்காட்டு", False, 'oorezhuthu'),
        ]
        statement, is_true, rule = random.choice(statements)
        return {
            'question': f"'{statement}' - சரியா? (y/n)",
            'answer': is_true,
            'explanation': f"இது {RULE_EXPLANATIONS[rule]['name']} விதியாகும்." if is_true else f"இது {RULE_EXPLANATIONS[rule]['name']} விதியில் வராது.",
            'type': 'true_false'
        }

    def generate_question_fill_blank(self):
        """வெற்றிட நிரப்பும் கேள்வி"""
        blanks = [
            ("_______ என்பது ஓரெழுத்தொருமொழிக்கு எடுத்துக்காட்டு", ['ஆ', 'ஈ', 'ஊ', 'ஏ', 'ஐ', 'ஓ'], 'oorezhuthu'),
            ("'அக்கொலை' என்பது _______ விதியில் வரும்", ['சுட்டெழுத்து', 'suttu'], 'suttu'),
            ("மெல்லின மெய் எழுத்துக்கள்: ஞ், ண், ந், ம், _______", ['ன்'], 'mellinam'),
            ("இடையின மெய் எழுத்துக்கள்: ய், ர், ல், வ், ழ், _______", ['ள்'], 'idaiyinam'),
        ]
        question, possible_answers, rule = random.choice(blanks)
        return {
            'question': question,
            'answer': possible_answers,
            'explanation': RULE_EXPLANATIONS[rule]['example'],
            'type': 'fill_blank'
        }

    def generate_question(self):
        """சீரற்ற கேள்வியை உருவாக்கு"""
        q_type = self.get_question_type()
        if q_type == 'identify_rule':
            return self.generate_question_identify_rule()
        elif q_type == 'find_example':
            return self.generate_question_find_example()
        elif q_type == 'true_false':
            return self.generate_question_true_false()
        else:
            return self.generate_question_fill_blank()

    def check_answer(self, question_obj, user_answer):
        """விடையைச் சரிபார்"""
        if question_obj['type'] == 'identify_rule':
            rule_keys = {v['name']: k for k, v in RULE_EXPLANATIONS.items()}
            rule_keys.update({k: k for k in RULE_EXPLANATIONS.keys()})

            user_clean = user_answer.strip().lower()
            expected = question_obj['answer']

            if user_clean in rule_keys:
                return rule_keys[user_clean] == expected
            return user_clean == expected or user_clean == RULE_EXPLANATIONS[expected]['name'].lower()

        elif question_obj['type'] == 'find_example':
            expected_rule = question_obj['answer']
            # எடுத்துக்காட்டு சரியான விதியில் உள்ளதா?
            rules_for_word = get_all_rules_for_word(user_answer)
            return expected_rule in rules_for_word

        elif question_obj['type'] == 'true_false':
            user_bool = user_answer.lower() in ['y', 'yes', 's', 'சரி', 'ஆம்']
            return user_bool == question_obj['answer']

        elif question_obj['type'] == 'fill_blank':
            user_clean = user_answer.strip()
            if isinstance(question_obj['answer'], list):
                for ans in question_obj['answer']:
                    if user_clean == ans or user_clean.lower() == ans.lower():
                        return True
                return False
            return user_clean == question_obj['answer']

        return False

    def get_correct_answer_text(self, question_obj):
        """சரியான விடையின் விளக்கத்தைப் பெறுக"""
        if question_obj['type'] == 'identify_rule':
            rule_key = question_obj['answer']
            return f"{RULE_EXPLANATIONS[rule_key]['name']} - {RULE_EXPLANATIONS[rule_key]['rule']}"
        elif question_obj['type'] == 'find_example':
            examples = EXAMPLE_WORDS.get(question_obj['answer'], ['தமிழ்ச்சொல்'])
            return f"எடுத்துக்காட்டு: {random.choice(examples)}"
        elif question_obj['type'] == 'true_false':
            return question_obj.get('explanation', 'தொல்காப்பிய விதிப்படி இது சரியான விடை')
        else:
            return question_obj.get('explanation', 'தொல்காப்பிய விதிப்படி இதுவே சரியான விடை')

    def display_question(self, question_obj, q_num):
        """கேள்வியைக் காட்டு"""
        print(f"\n{'='*60}")
        print(f"📝 கேள்வி {q_num}/15")
        print(f"{'='*60}")

        if question_obj['type'] == 'identify_rule':
            print(f"\n🔍 {question_obj['question']}\n")
            print("💡 விதிகள்: உயிர்|மெல்லினம்|இடையினம்|ஓரெழுத்தொருமொழி|அளபெடை|சுட்டெழுத்து|வினாச்சொல்")

        elif question_obj['type'] == 'find_example':
            print(f"\n✍️ {question_obj['question']}\n")
            print(f"📌 {RULE_EXPLANATIONS[question_obj['rule_key']]['rule']}")

        elif question_obj['type'] == 'true_false':
            print(f"\n❓ {question_obj['question']}\n")

        else:
            print(f"\n📖 {question_obj['question']}\n")

    def play_round(self, q_num):
        """ஒரு சுற்று விளையாடு"""
        question_obj = self.generate_question()
        self.display_question(question_obj, q_num)

        # உதவி விருப்பம்
        print("\n[Enter] விடை தர | [h] உதவி/குறிப்பு | [s] விடை தெரியவில்லை")
        choice = input("\nதேர்வு: ").strip().lower()

        if choice == 'h':
            print(f"\n💡 {self.get_correct_answer_text(question_obj)[:100]}...")
            print("\nதொல்காப்பியம்: எழுத்ததிகாரத்தை நோக்குக!")
            user_answer = input("\nஇப்போது உங்கள் விடை: ").strip()
        elif choice == 's':
            print(f"\n📖 சரியான விடை: {self.get_correct_answer_text(question_obj)}")
            print("\n🌟 தொடர்ந்து கற்போம்!")
            self.streak = 0
            return False
        else:
            user_answer = choice if choice else input("\nஉங்கள் விடை: ").strip()

        if self.check_answer(question_obj, user_answer):
            points = 10 + (self.streak * 2)  # Streak bonus
            print(f"\n✅ சரியான விடை! +{points} புள்ளிகள்!")
            if self.streak > 0:
                print(f"🔥 {self.streak} சரியான பதில்களுக்கான போனஸ்!")

            self.score += points
            self.correct_answers += 1
            self.streak += 1
            self.max_streak = max(self.max_streak, self.streak)

            # விளக்கங்கள்
            print(f"\n📚 {self.get_correct_answer_text(question_obj)}")
            return True
        else:
            print(f"\n❌ முறையான விடை அல்ல.")
            print(f"📖 சரியான விடை: {self.get_correct_answer_text(question_obj)}")
            self.streak = 0
            return False

    def display_progress(self):
        """முன்னேற்றத்தைக் காட்டு"""
        print(f"\n{'─'*50}")
        print(f"🏆 மதிப்பெண்: {self.score}")
        print(f"📊 கேள்விகள்: {self.total_questions}")
        if self.total_questions > 0:
            accuracy = (self.correct_answers / self.total_questions) * 100
            print(f"📈 துல்லியம்: {accuracy:.1f}%")
        print(f"🔥 அதிகபட்ச தொடர்: {self.max_streak}")
        print(f"{'─'*50}")

    def display_certificate(self):
        """சான்றிதழ் காட்டு"""
        accuracy = (self.correct_answers / self.total_questions) * 100 if self.total_questions > 0 else 0

        print("\n" + "="*70)
        print("🏅 தொல்காப்பிய மொழியிறுதி விதிகள் - தேர்ச்சிச் சான்றிதழ் 🏅")
        print("="*70)
        print(f"\n   மாணவர் பெயர்: தமிழிலக்கிய ஆர்வலர்")
        print(f"   மொத்த மதிப்பெண்: {self.score}")
        print(f"   சரியான விடைகள்: {self.correct_answers}/{self.total_questions}")
        print(f"   துல்லியம்: {accuracy:.1f}%")
        print(f"   அதிகபட்ச தொடர்: {self.max_streak}")

        if accuracy >= 80:
            print("\n   📜 தரம்: முதல் வகுப்பு (Distinction)")
            print("   🌟 சிறந்த முறையில் தொல்காப்பிய விதிகளைக் கற்றுள்ளீர்கள்!")
        elif accuracy >= 60:
            print("\n   📜 தரம்: இரண்டாம் வகுப்பு (First Class)")
            print("   📖 நல்ல முயற்சி! தொடர்ந்து பயின்று மேலும் சிறக்க!")
        elif accuracy >= 40:
            print("\n   📜 தரம்: மூன்றாம் வகுப்பு (Second Class)")
            print("   💪 மேலும் பயிற்சி தேவை. விதிகளை மீண்டும் கற்கவும்!")
        else:
            print("\n   📜 தரம்: தேர்ச்சி (Pass)")
            print("   📚 விதிகளை மீண்டும் படித்து முயற்சிக்கவும்!")

        print("\n" + "="*70)
        print("🙏 தொல்காப்பியரின் தமிழிலக்கணம் போற்றிப் புகழ்வோம்!")
        print("   வாழ்க தமிழ்! வளர்க தமிழிலக்கியம்!")
        print("="*70)

    def run(self):
        """முதன்மை விளையாட்டு இயக்கி"""
        self.display_header()

        # விதிகளைக் காட்டவா?
        print("\n📖 தொல்காப்பிய விதிகளை முதலில் பார்க்கவா? (ஆம் y/இல்லை n)")
        if input().lower() in ['y', 'yes', 's', 'சரி']:
            self.display_rules_summary()

        print("\n" + "="*50)
        print("🎮 விளையாட்டு வழிகாட்டி:")
        print("   15 கேள்விகள் | தொடர்ச்சியான சரியான பதில்களுக்கு போனஸ் புள்ளிகள்")
        print("   'h' - உதவி குறிப்பு | 's' - விடை தெரியவில்லை")
        print("="*50)

        input("\n✨ விளையாட்டைத் தொடங்க Enter அழுத்தவும்...")

        # 15 கேள்விகள்
        for round_num in range(1, 16):
            self.total_questions += 1
            self.play_round(round_num)
            self.display_progress()

            if round_num < 15:
                print("\n▶️ அடுத்த கேள்விக்கு Enter... (q - வெளியேற)")
                if input().lower() == 'q':
                    break

        self.display_certificate()

# ============================================================
# நிரலை இயக்கும் இடம்
# ============================================================

def main():
    """முதன்மைச் செயல்பாடு"""
    game = TholkaappiyamGame()
    try:
        game.run()
    except KeyboardInterrupt:
        print("\n\n👋 விடைபெறுகிறேன். நன்றி!")
        print("🙏 தொடர்ந்து தொல்காப்பியம் கற்க வாழ்த்துகள்!")
        sys.exit(0)
    except ImportError:
        print("\n⚠️ 'tamil' நூலகம் நிறுவப்படவில்லை!")
        print("💡 pip install tamil என நிறுவவும்")
        sys.exit(1)
    except Exception as e:
        print(f"\n⚠️ பிழை: {e}")
        sys.exit(1)

if __name__ == "__main__":
    main()
```
